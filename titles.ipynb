```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('netflix_titles.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.drop(df.columns[12:], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df['duration'].describe()
```

```python
# Histograma
plt.figure(figsize=(10, 5))
sns.histplot(df['duration'], bins=10, kde=True)
plt.title('Distribución de la duración de títulos')
plt.xlabel('Duración ')
plt.ylabel('Frecuencia')
plt.show()

# Box plot
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['duration'])
plt.title('Box plot de la duración de títulos')
plt.xlabel('Duración ')
plt.show()
```

```python
lanzamientos=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(8,6))
plt.plot(lanzamientos.index,lanzamientos.values,color='Red')
plt.title('Distribucion de lanzamientos por año')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
